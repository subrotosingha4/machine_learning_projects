# Imports

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

# Model

```python
model = Sequential([Dense(units=1, input_shape=[1])])
```

# Model compile

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

# Data

- Data relation: Y = 2X â€“ 1

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

# Model Training

```python
model.fit(xs, ys, epochs=500)
```

# Prediction

```python
print(model.predict([10.0]))
```

# What Network Learned

```python
Layer_0 = Dense(units=1, input_shape=[1])
model = Sequential([Layer_0])
model.compile(optimizer='sgd', loss='mean_squared_error')

xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

model.fit(xs, ys, epochs=500, verbose= False)

print(model.predict([10.0]))
print("Here is what the model learned: {}".format(Layer_0.get_weights()))
```

 - The learned relationship between X and Y was Y = 1.9967953X â€“ 0.9900647
