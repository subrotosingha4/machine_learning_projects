Solving methodology:
- firstly, this is an unsupervised problem since no data is labeled.
- so to detect the anomaly, an unsupervised machine learning algorithm will be trained on this dataset and later will be tested to determine any anomalies
- since there is no aproriori knowledge about an unsupervised algorithm which would work best thus 12 machine learnign algorithms will be trained using PyCaret 
- the randomly two popular unsupervised machine learning algorithm will be used for anomaly detection
- the two algorithms are 'knn' (K-Nearest Neighbors Detector) and 'lor' (Local Outlier Factor	)
-  'lor'  which computes the local density deviation of a given data point with respect to its neighbors and detect outliers
- 'knn' works by finding the distances between a query and all the examples in the data
- finally, the given data [30,22] was tested and found that this is an anomaly with respect the given trained data

```python
from pycaret.anomaly import *
```

```python
setup = setup(df, session_id = 123, numeric_features=['Width','Height'],normalize=True)
```

```python
models()
```

# knn

```python
knn = create_model('knn')
```

```python
knn_results = assign_model(knn)
knn_results.head()
```

```python
d = {'Width': [30], 'Height': [22]}
data_unseen=pd.DataFrame(data=d)
```

```python
data_unseen
```

# predict [30,22]

```python
knn_predictions = predict_model(model = knn, data = data_unseen)
knn_predictions
```

- anomaly
